```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


social_data = pd.read_excel("Dataset.xlsx", sheet_name="Social Listening_Data",skiprows=5)


selected_columns = ['Date', 'Social Mention/Comment', 'Sentiment','Emotion']


celebration_mentions = social_data[social_data['Social Mention/Comment'].str.contains('celebrate', case=False, na=False)]
celebration_mentions = celebration_mentions[selected_columns].dropna()
print("\nMentions related to celebrations:")
print(celebration_mentions)


holiday_mentions = social_data[social_data['Hashtags'].str.contains('holiday', case=False, na=False)]
holiday_mentions = holiday_mentions[selected_columns].dropna()
print("\nMentions related to holidays:")
print(holiday_mentions)


joyful_mentions = social_data[social_data['Emotion'].str.contains('joy', case=False, na=False)]
joyful_mentions = joyful_mentions[selected_columns].dropna()

print("\nJoyful mentions:")
print(joyful_mentions)
```

```python

emotion_counts = joyful_mentions['Sentiment'].value_counts()

# Pie chart
plt.figure(figsize=(5, 5))
plt.pie(emotion_counts, labels=emotion_counts.index, autopct='%1.1f%%', colors=sns.color_palette('Set3'), startangle=90)
plt.title('Distribution of Joyful Mentions by Emotion')
plt.show()
```

```python
emotion_counts = joyful_mentions['Sentiment'].value_counts()
plt.figure(figsize=(5, 5))
sns.barplot(x=emotion_counts.index, y=emotion_counts.values, palette='Set3', saturation = 0.1)
plt.title('Distribution of Joyful Mentions by Emotion')
plt.xlabel('Sentiment')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')  
plt.show()
```
