```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


social_data = pd.read_excel("Dataset.xlsx", sheet_name="Social Listening_Data", skiprows=5)
search_data = pd.read_excel("Dataset.xlsx", sheet_name="Google Search_Data", skiprows = 7)
reviews_data = pd.read_excel("Dataset.xlsx", sheet_name="E-Commerce Reviews_Data", skiprows=5)
brand_data = pd.read_excel("Dataset.xlsx", sheet_name="Brand Tracking_Data", skiprows=11)

sentiment_distribution = social_data['Sentiment'].value_counts()
print("Sentiment Distribution in Social Media Mentions:")
print(sentiment_distribution)

```

```python
import matplotlib.pyplot as plt
import seaborn as sns


labels = social_data['Sentiment'].value_counts().index
sizes = social_data['Sentiment'].value_counts().values


colors = sns.color_palette("viridis")

plt.figure(figsize=(5, 3))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', colors=colors, startangle=140)
plt.title("Sentiment Distribution for Leffe Mentions")
plt.show()
```

```python
# Plot Sentiment Distribution
plt.figure(figsize=(8, 6))
sns.countplot(x='Sentiment', data=social_data, palette="viridis")
plt.title("Sentiment Distribution for Leffe Mentions")
plt.xlabel("Sentiment")
plt.ylabel("Count")
plt.show()

```
